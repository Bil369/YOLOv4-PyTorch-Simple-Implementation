```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
total_loss = pd.read_csv('total_loss.csv', header=None)
```

```python
plt.figure(dpi=300)
plt.title('Train loss')
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.grid()
plt.plot(total_loss, linewidth=3, marker='o')
plt.savefig('total_loss.png')
```
